```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.preprocessing import PolynomialFeatures
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
import warnings
warnings.filterwarnings('ignore')
pd.set_option('display.max_columns', 500)
```

```python
df1 = pd.read_csv('../data/df_transform_zeroes_excluded.csv')
df1.drop('Unnamed: 0', axis=1, inplace=True)
df1.head()
```

```python
np.abs(df1.corr()['Response']).sort_values(ascending=False)
```

```python
# dropping columns I should have before based on EDA, low correlation, and prevention of multicolinearity/redundancy
df1 = df1.drop(['marital_status_Together','NumWebVisitsMonth','MntFruit_PerIncome','Complain','Education_2n Cycle',
                'Dt_Custom_Weekday_4','MntSweet_PerIncome','Dt_Customer_Season_Winter','NumDealsPurchases'], axis=1)
```

```python
# creating dummies for technically continuous variables, but ones that only go from 0 to 2
kid_dummies = pd.get_dummies(df1.Kidhome, prefix='Kidhome',drop_first=True)
teen_dummies = pd.get_dummies(df1.Teenhome, prefix='Teenhome',drop_first=True)
df1 = df1.drop(['Kidhome','Teenhome'], axis=1)
df1 = pd.concat([df1,kid_dummies,teen_dummies],axis=1)
```

```python
sns.set_style('darkgrid')
sns.countplot(df1.Response)
```

```python
# splitting the data
X = df1.loc[:,df1.columns!='Response']
y = df1.Response
```

```python
# attempted to use polynomial features but didn't make much of a difference in the model
# pn=PolynomialFeatures(degree=2,include_bias=False)
# pn.fit(X)
# columns=pn.get_feature_names(X.columns)
# poly_X=pd.DataFrame(pn.transform(X), columns=columns)
# poly_X.shape
```

```python
# train test split!
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=15)
```

```python
# scaling
scale = MinMaxScaler()
scale.fit(X_train)
X_train = scale.transform(X_train)
X_test = scale.transform(X_test)
```

```python
# accounting for class imbalance
from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=15)
X_train,y_train = sm.fit_sample(X_train,y_train)
```

```python
# logistic regression baseline
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression(random_state=15)
logreg.fit(X_train,y_train)
y_pred_train = logreg.predict(X_train)
y_pred_test = logreg.predict(X_test)
```

```python
from tqdm import tqdm_notebook as tq
acc_scores_train = []
acc_scores_test = []
for x in tq(range(100,1600,100)):
    xgbc = XGBClassifier(random_state=15,n_estimators=x, max_depth=6)
    xgbc.fit(X_train,y_train)
    y_pred_train = xgbc.predict(X_train)
    y_pred_test = xgbc.predict(X_test)
    acc_scores_train.append(metrics.accuracy_score(y_train,y_pred_train))
    acc_scores_test.append(metrics.accuracy_score(y_test,y_pred_test))
f, (ax1,ax2) = plt.subplots(1,2,figsize=(18,6))
ax1.plot(range(100,1600,100), acc_scores_train)
ax2.plot(range(100,1600,100), acc_scores_test)
```

```python
def frange(x, y, jump):
    while x < y:
        yield x
        x += jump
learning_rate = list(frange(.05,1,.05))
learning_rate
```
